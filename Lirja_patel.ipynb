```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("tips.csv")
```

```python
data.head()
```

```python
data
```

```python
#Dropping useless columns
data=data.drop("smoker",axis=1)
```

```python
data
```

```python

data=data.drop("Payer Name",axis=1)
```

```python
data=data.drop("CC Number",axis=1)
data=data.drop("Payment ID",axis=1)
```

```python
data
```

```python
# graph plotting
df = pd.DataFrame(data)

# Plot 1: Total Bill vs Tip
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex',  sizes=(20, 200))
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Attributes')
plt.show()
```

```python
# Plot 2: Price Per Person vs Tip
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price_per_person', y='tip', data=df, hue='size',  sizes=(20, 200))
plt.title('Price Per Person vs Tip')
plt.xlabel('Price Per Person ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Attributes')
plt.show()
```

```python
# Plot 3: Tip Distribution by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip ($)')
plt.show()
```

```python
# Plot 4: Pie Chart of Gender Distribution
plt.figure(figsize=(8, 8))
gender_counts = df['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

```python
# Plot 5: Histogram of Total Bill
plt.figure(figsize=(10, 6))
sns.histplot(df['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Converting String to Integer, Label Encoding
from sklearn.preprocessing import LabelEncoder


label_encoder = LabelEncoder()
data['sex_encoded'] = label_encoder.fit_transform(data['sex'])#Female 0 , male 1
```

```python

data['time_encoded'] = label_encoder.fit_transform(data['time'])

data['day_encoded'] = label_encoder.fit_transform(data['day'])
```

```python
data
```

```python
x = data[['total_bill','sex_encoded','day_encoded','time_encoded','size','price_per_person']]
y = data[['tip']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test= train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
import math
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse= math.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"root mean Squared Error: {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```
